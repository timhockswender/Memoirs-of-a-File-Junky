```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#draw some Limacons
theta = np.linspace(0, 2.*np.pi, 200)
fig, (ax1, ax2, ax3) = plt.subplots(1,3,subplot_kw={'projection': 'polar'} , figsize = (10, 4) )
fig.subplots_adjust(wspace=1, hspace=10)
#fig.tight_layout()
# Plot each Limaçon
a=2; b=3
r1 = a + b*np.cos(theta)
ax1.plot(theta+(r1<0)*np.pi, np.abs(r1), 'r')  #Clever trick
a=2; b=2
r2 = a + b*np.cos(theta)
ax2.plot(theta+(r2<0)*np.pi, np.abs(r2), 'g')  #Clever trick
a=3; b=2
r3 = a + b*np.cos(theta)
ax3.plot(theta+(r3<0)*np.pi, np.abs(r3), 'b')  #Clever trick
# Create a legend
ax2.set_title(f'Cosine Based Limacon with Various Parameters ', loc  = 'center') #also 'left' : 'right'
plt.show()
```
