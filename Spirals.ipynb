```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Archimedes Spiral
size = 6
theta = np.linspace(0, size*np.pi, 200) 
radius =theta/2
fig, (ax1) = plt.subplots(1,1,subplot_kw={'projection': 'polar'} , figsize = (4, 4 )  )
#ax.set_rticks([ 1,  2, 3, 4.])  # fewer radial ticks works ok iff size=4
ax1.plot(theta+(radius<0)*np.pi, np.abs(radius), 'b')  #Clever trick
ax1.set_yticklabels([]) #optional
ax1.set_title(f'Archimedes Spiral With Size = {size}', loc  = 'center') #also 'left' : 'right'
plt.show()
```

```python
#Fermats Spiral - HALF
size = 4
theta = np.linspace(0, 8*np.pi, 200) 
#theta = np.linspace(-8*np.pi, 8*np.pi, 200) 
radius = np.sqrt(theta)
fig, (ax1) = plt.subplots(1,1,subplot_kw={'projection': 'polar'} , figsize = (4, 4 )  )
ax1.set_rticks([ ])  
ax1.plot(theta+(radius<0)*np.pi, np.abs(radius), 'b')  #Clever trick
ax1.set_yticklabels([]) #optional
ax1.set_title(f'Fermat Spiral With Size = {size}', loc  = 'center') #also 'left' : 'right'
plt.show()
```

```python
def fermat_spiral(theta, a=1):
  """
  This function defines Fermat's spiral in polar coordinates.

  Args:
      theta: Angle in radians.
      a: Positive constant scaling the spiral (default 1).

  Returns:
      Radius (distance from origin) at the given angle.
  """
  return a * np.sqrt(theta)

theta = np.linspace(0, 100, 1000)  # Adjust range and number of points as needed
# Create the polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Plot the spiral using the function and theta values
ax.plot(theta, fermat_spiral(theta))

# Set plot options (optional)
ax.set_title("Fermat's Spiral")

plt.show()
```

```python
def fermat_spiral_clockwise(theta, a=1):
  """
  This function defines the clockwise arm of Fermat's spiral in polar coordinates.

  Args:
      theta: Angle in radians (negative values for clockwise).
      a: Positive constant scaling the spiral (default 1).

  Returns:
      Radius (distance from origin) at the given angle.
  """
  return a * np.sqrt(-theta)  # Negative theta for clockwise
```

```python
# Generate negative theta values for clockwise arm
theta_cw = np.linspace(-100, 0, 1000)  # Adjust range and number of points

# Plot counter-clockwise arm
ax.plot(theta, fermat_spiral(theta))

# Plot clockwise arm (using the separate function)
ax.plot(theta_cw, fermat_spiral_clockwise(theta_cw))
plt.show()
```

```python
# Combine theta values for both arms (ensure negative values are included)
theta = np.concatenate((np.linspace(-100, 0, 500), np.linspace(0, 100, 500)))

# Plot using a single theta array
ax.plot(theta, fermat_spiral(np.abs(theta)))  # Use absolute value for radius
plt.show()
```

```python

def fermat_spiral(theta, a=1):
  """
  This function defines Fermat's spiral in polar coordinates.

  Args:
      theta: Angle in radians.
      a: Positive constant scaling the spiral (default 1).

  Returns:
      Radius (distance from origin) at the given angle.
  """
  return a * np.sqrt(np.abs(theta))  # Ensure positive radius for both arms

# Combine theta values for both arms
theta = np.concatenate((np.linspace(-100, 0, 500), np.linspace(0, 100, 500)))  #original
theta = np.concatenate((np.linspace(-30, 0, 500), np.linspace(0, 30, 500))) #ugly

# Create the polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Plot the spiral using the combined theta and absolute value for radius
ax.plot(theta, fermat_spiral(theta))

# Set plot options (optional)
ax.set_title("Fermat's Spiral (Both Arms)")

plt.show()
```
