```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate theta values
theta = np.linspace(0, 2*np.pi, 200)  #200 point seems generall enough but increase in some plots to improve smoothnes
```

```python
#a couple simple functions
def rose_sine(size, lobes, theta):
    return size * np.sin(lobes * theta )
def rose_cosine(size, lobes, theta):
    return size * np.cos(lobes * theta )    
```

```python
#Four polar roses
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, subplot_kw=dict(projection='polar'), figsize=(10, 4))
fig.tight_layout()
fig.suptitle('Rose Plots Using Sines; Different Lobes')
size=4
lobes=2
r1 = rose_sine(size, lobes,theta)
lobes = 3
r2 = rose_sine(size, lobes,theta)
lobes = 4
r3 = rose_sine(size, lobes,theta)
lobes = 5
r4 = rose_sine(size, lobes,theta)
ax1.plot(theta+(r1<0)*np.pi, np.abs(r1), 'r')  #Clever trick
ax2.plot(theta+(r2<0)*np.pi, np.abs(r2), 'y')  #Clever trick
ax3.plot(theta+(r3<0)*np.pi, np.abs(r3), 'b')  #Clever trick
ax4.plot(theta+(r4<0)*np.pi, np.abs(r4), 'g')  #Clever trick
plt.show()  #Avoids text output in Jupyter Lab
```

```python
#single polar rose plot as demo
theta = np.linspace(0, 2*np.pi, 200)
size = 4
lobe_factor =4
radius = rose_sine(size, lobe_factor,theta)
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.set_rticks([ 1,  2, 3, 4.])  # fewer radial ticks works ok iff size=4
ax.plot(theta+(radius<0)*np.pi, np.abs(radius), 'b', f'a={size}, b={lobe_factor}')  #Clever trick
ax.set_xlabel('X Label 4')
plt.title(f'Sine Rose With {lobes} * sin$\\theta$; size = {size}', loc  = 'center') #also 'left' : 'right'

plt.show()
```

```python
# Create a polar plot with a little embellishment
theta = np.linspace(0, 2*np.pi, 200)
size = 4
radius = rose_sine(4,2,theta)
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'), facecolor='lightskyblue',layout='constrained' )
ax.plot(theta+(radius<0)*np.pi, np.abs(radius), ' g')  #Clever trick
plt.title(f'Sine Rose With 2 * sin(theta); size = {size}', loc  = 'center') #also 'left' : 'right'
ax.set_rticks([ 1,  2, 3, 4.])  # fewer radial ticks works ok iff size=4
plt.show()
```
