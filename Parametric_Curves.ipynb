```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#an approach to parametric functions
# Define the function for a limacon (via Wolfram)
def f(x, y):
  return x**4 + 2*x**2*y**2 + y**4 + 6*y*(x**2 + y**2) - 4*x**2 + 5*y**2

# Create a mesh grid
x = np.linspace(-5, 5, 200)
y = np.linspace(-6, 2.5, 200)
X, Y = np.meshgrid(x, y)

# Evaluate the function on the mesh grid
Z = f(X, Y)

# Create the contour plot
plt.contour(X, Y, Z, levels=20) #20 is standars

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Contour Plot of the Function")
plt.grid()
# Show the plot
plt.show()
```

```python
#an approach to parametric functions
#looking at x and y ranges; may affect symmetry
# Define the function for a limacon (via Wolfram)
def f(x, y):
  return x**4 + 2*x**2*y**2 + y**4 + 6*y*(x**2 + y**2) - 4*x**2 + 5*y**2

# Create a mesh grid
x = np.linspace(-3.2, 3.2, 200)  #choose x range to suit figure
y = np.linspace(-5, 1.0, 200)  #choose y range to suit figure
X, Y = np.meshgrid(x, y)

# Evaluate the function on the mesh grid
Z = f(X, Y)
fig, (ax1) = plt.subplots(1,1, figsize = (6, 4) )
# Create the contour plot - adjusted to look like a simple limcon
ax1.contour(X, Y, Z, levels=1) #20 is standard

# Add labels and title
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_title("Zoomed in to a Simple, Slightly Flattened, Limacon")
ax1.grid()
# Show the plot
plt.show()
```

```python
#draw some unusual figures
#NOTE: the b parameter will multiply the angle, not the sine of the angle
    #the b parameter does not change, only the domain of the function changes
theta = np.linspace(-9/2*np.pi, 9/2*np.pi, 2000)
fig, (ax1, ax2, ax3) = plt.subplots(1,3,subplot_kw={'projection': 'polar'} , figsize = (12, 4) )
fig.subplots_adjust(wspace=1, hspace=10)
#fig.tight_layout()
# Plot each figure
a=0; b=4/9
#r1 = a + b*np.sin(b*theta)
r1 = a + np.sin(b*theta)
ax1.plot(theta+(r1<0)*np.pi, np.abs(r1), 'r')  #Clever trick
ax1.set_yticklabels([])
theta = np.linspace(0, 9*np.pi, 2000)
r2 = a + np.sin(b*theta)
ax2.plot(theta+(r2<0)*np.pi, np.abs(r2), 'g')  #Clever trick
ax2.set_yticklabels([])
theta = np.linspace(0, 18*np.pi, 2000)
r3 = a + np.sin(b* theta)
ax3.plot(theta+(r3<0)*np.pi, np.abs(r3), 'b')  #Clever trick
ax3.set_yticklabels([])
# Create a legend
ax2.set_title(f'Sine Based Figures with Various Parameters ', loc  = 'center') #also 'left' : 'right'
#clean it up
ax1.set_yticklabels([])
plt.show()
```
